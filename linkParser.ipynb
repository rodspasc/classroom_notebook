```python
from urllib.request import urlopen
from html.parser import HTMLParser

class linkParser(HTMLParser):
    '''analisador de doc. HTML que mostra valores dos
atributos href nas tags de incio de ncora'''
    
    def handle_starttag(self,tag,attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    print(attr[1])
                    
def getSource(url):
    response = urlopen(url)
    html = response.read()
    return html.decode()

html = getSource('http://www.uol.com.br')
parser = linkParser()
parser.feed(html)
```
